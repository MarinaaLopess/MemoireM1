```python
df_merged = pd.merge(df, df_entities, on="Ouvrage", suffixes=("_tokens", "_ents"))

ratio_data = {
    "Ouvrage": df_merged["Ouvrage"],
    "spaCy (sm)": df_merged["spaCy (sm)_ents"] / df_merged["spaCy (sm)_tokens"],
    "spaCy (lg)": df_merged["spaCy (lg)_ents"] / df_merged["spaCy (lg)_tokens"],
    "Flaubert": df_merged["Flaubert_ents"] / df_merged["Flaubert_tokens"],
    "Camembert": df_merged["Camembert_ents"] / df_merged["Camembert_tokens"],
    "Stanza": pd.to_numeric(df_merged["Stanza_ents"], errors="coerce") / df_merged["Stanza_tokens"],
}

df_ratios = pd.DataFrame(ratio_data)

from tabulate import tabulate
print(tabulate(df_ratios, headers='keys', tablefmt='github', showindex=False))
```

```python
#Generation du tableau de la densité par ouvrage sur Latex

latex_code = r"""
\begin{table}[ht]
\centering
    \caption{\small DENSITÉ D'ENTITÉS NOMMÉES (\%) PAR OUVRAGE}
    \captionsetup{skip=0.5cm}
    \begin{tblr}{
      colspec={@{}X[4,l,font=\bfseries]X[c]X[c]X[c]X[c]X[c]},
      row{1}={font=\itshape}
    }
        \hline
        \textbf{Ouvrage} & \textbf{Densité spaCy (sm)} & \textbf{Densité spaCy (lg)} & \textbf{Densité Flaubert} & \textbf{Densité Camembert} & \textbf{Densité Stanza} \\
        \hline
"""

for idx, row in df_ratios.iterrows():
    ouvrage = row['Ouvrage'].replace('&', '\\&')  # Échapper les caractères spéciaux
    # Multiplier par 100 pour obtenir un pourcentage
    line = f"{ouvrage} & {row['spaCy (sm)']*100:.2f}\\% & {row['spaCy (lg)']*100:.2f}\\% & {row['Flaubert']*100:.2f}\\% & {row['Camembert']*100:.2f}\\% & {row['Stanza']*100:.2f}\\% \\\\"
    latex_code += line + "\n\\hline\n"

latex_code += r"""
    \end{tblr}
\end{table}
"""

print(latex_code)

with open("densite_tableau.tex", "w", encoding="utf-8") as f:
    f.write(latex_code)
```
